```python
APA = stock_price[(stock_price['Symbol']=='APA')]
APA
```

```python
import plotly.graph_objects as go
import pandas as pd


df = APA
fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```
